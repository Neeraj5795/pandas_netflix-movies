```python
# In which year highest no of movies  were released

movie=data[(data["Category"]=="Movie")]
gp = movie.groupby("Year").agg({"Category":"count"}).sort_values(by="Category",ascending=False).head(1)
gp
```

```python
# Most popular movie in 2019
movie=data[(data["Category"]=="Movie") & (data["Year"]==2019)]
gp=movie.groupby("Title").agg({"Imdb Score":"max"}).sort_values(by="Imdb Score",ascending=False).head(1)
gp
```

```python
# in how many movies and tv shows Tom Cruise was cast
dt=data[data["Cast"].str.contains('Tom Cruise')]
dt["Category"].count()

```

```python
# how many movies got TV-14 rating in canada
dt=data[(data["Category"]=="Movie") & (data["Rating"]=="TV-14") & (data["Production Country"]=="Canada")]
dt["Show Id"].count()
```

```python
# what is the maximum duration of movies & tv show on netflix
data[["Minute","Unit"]]=data["Duration"].str.split(" ",expand=True)
data["Minute"]=data["Minute"].astype("int64")
data["Minute"].max()
```

```python
# which individual country has the highest no of tv shows
dt=data[(data["Category"]=="TV Show")]
dt.groupby("Production Country").agg({"Category":"count"}).sort_values(by="Category",ascending=False).head(1)
```
